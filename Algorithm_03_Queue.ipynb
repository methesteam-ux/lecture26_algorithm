```python
queue = []
capacity = 5

def isFull():
    if len(queue) >= capacity:
        return True
    else:
        return False
    
def isEmpty():
    if len(queue) == 0:
        return True
    else:
        return False

def enqueue(data):
    if not isFull():
        queue.append(data)
    else:
        print("Queue가 차 있어서 더 이상 추가할 수 없습니다.")

def dequeue():
    if not isEmpty():
        print(f"> 가져온 데이터 : {queue.pop(0)}")
    else:
        print("Queue가 비어 있습니다.")

def getFront():
    if not isEmpty():
        print(f"> 가져올 데이터 : {queue[0]}")
    else:
        print("Queue가 비어 있습니다.")

while True:
    # 메뉴 출력
    print("===================================================")
    print("  1.Enqueue    2.Dequeue    3.Get Front    0.Exit  ")
    print("===================================================")
    menu = int(input(">> 연산 메뉴 선택 :"))
    #print(menu)
    
    if menu == 0: 
        break
    elif menu == 1: # Enqueue
        data = int(input(">> 추가할 데이터 입력 :"))
        enqueue(data)
    elif menu == 2: # Dequeue
        dequeue()
    elif menu == 3: # Get Front
        getFront()
    
    print("[Queue]", queue)

print("정수형 Queue 종료")
```
