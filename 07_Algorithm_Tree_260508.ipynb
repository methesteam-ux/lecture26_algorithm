### 1. 이진 트리 (Binary Tree)

```python
class TreeNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

    def set_left(self, left):
        self.left = left

    def set_right(self, right):
        self.right = right
```

```python
my_tree = TreeNode("A", None, None)

new_node = TreeNode("B", None, None)
my_tree.set_left(new_node)

new_node = TreeNode("C", None, None)
my_tree.set_right(new_node)

cur_node = my_tree.left # node B 객체 (주소)
cur_node.set_left(TreeNode("D", None, None))
cur_node.set_right(TreeNode("E", None, None))

cur_node = my_tree.right
```

### 클래스로 이진 트리 구현

```python
d = TreeNode("D", None, None)
e = TreeNode("E", None, None)
b = TreeNode("B", d, e)

f = TreeNode("F", None, None)
c = TreeNode("C", f, None)

a = TreeNode("A", b, c)

my_binary_tree = a
```

```python
# preorder (선위 순회)
def preorder(node):
    if node == None: return

    print(node.data, end=" ")
    preorder(node.left)
    preorder(node.right)

preorder(my_binary_tree)
```

```python
# inorder (중위 순회)
def inorder(node):
    if node == None: return

    inorder(node.left)
    print(node.data, end=" ")
    inorder(node.right)

inorder(my_binary_tree)
```

```python
# postorder (후위 순회)
def postorder(node):
    if node == None: return

    postorder(node.left)
    postorder(node.right)
    print(node.data, end=" ")

postorder(my_binary_tree)
```

### 딕셔너리로 이진 트리 구현

```python
tree = {}

tree["A"] = ["B", "C"]
tree["B"] = ["D", "E"]
tree["C"] = ["F", None]
tree["D"] = [None, None]
tree["E"] = [None, None]
tree["F"] = [None, None]

tree_root = "A"
```

```python
# preorder (선위 순회)
def preorder(data):
    if data == None: return

    print(data, end=" ")
    preorder(tree[data][0])
    preorder(tree[data][1])

preorder(tree_root)
```

```python
# inorder (중위 순회)
def preorder(data):
    if data == None: return

    preorder(tree[data][0])
    print(data, end=" ")
    preorder(tree[data][1])

preorder(tree_root)
```

```python
# postorder (후위 순회)
def preorder(data):
    if data == None: return

    preorder(tree[data][0])
    preorder(tree[data][1])
    print(data, end=" ")

preorder(tree_root)
```

### Binary Search Tree

```python
class STreeNode:
    def __init__(self, key, left, right):
        self.key = key
        self.left = left
        self.right = right

    def set_left(self, left):
        self.left = left

    def set_right(self, right):
        self.right = right
```

```python
# BST 생성
t3 = STreeNode(3, None, None)
t12 = STreeNode(12, None, None)
t7 = STreeNode(7, t3, t12)
t27 = STreeNode(27, None, None)
t31 = STreeNode(31, t27, None)
t26 = STreeNode(26, None, t31)
t18 = STreeNode(18, t7, t26)

my_bst = t18
```

```python
def search_bst(n, key):
    if n == None:
        return None
    elif key == n.key:
        return n
    elif key < n.key:
        return search_bst(n.left, key)
    else:
        return search_bst(n.right, key)
    
result = search_bst(my_bst, 12)
if result: 
    print(result.key)
else:
    print("없는 key입니다.")
```

```python
def search_bst_iter(n, key):
    while n != None:
        if key == n.key:
            return n
        elif key < n.key:
            n = n.left
        else:
            n = n.right
    return None

result = search_bst_iter(my_bst, 25)
if result: 
    print(result.key)
else:
    print("없는 key입니다.")
```

```python
# class STreeNode로 숫자를 순서대로 입력받아 binary search tree를 구성하기 : insert_bst() 함수 활용 (iteration으로 코딩)
def insert_bst(n, key):
    new_node = STreeNode(key, None, None)

    while True:
        if key < n.key:
            if n.left is None:
                n.set_left(new_node)
                break
            else:
                n = n.left
        else:
            if n.right is None:
                n.set_right(new_node)
                break
            else:
                n = n.right
    return new_node

def preorder(node):
    if node == None: return

    print(node.key, end=" ")
    preorder(node.left)
    preorder(node.right)

keys = [18, 7, 26, 3, 12, 27, 31]

my_bst = STreeNode(keys[0], None, None)
for num in keys[1:]:
    insert_bst(my_bst, num)

preorder(my_bst)
```
