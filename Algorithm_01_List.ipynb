```python
# 리스트와 반복구조
# 주어진 데이터가 어디에 저장되어 있는지 출력
data_list = [23, 45, 21, 14, 2, 67]
target = 14

index = 0
while index <= 5 and data_list[index] != target:
    index += 1

print(index)
```

```python
for index in range(6):
    if data_list[index] == target:
        break

print(index)
```

```python
# 리스트에 저장된 숫자 중 가장 큰 숫자 출력 (최댓값)
data_list = [123, 45, 10, 14, 2, 67]

max = data_list[0]
max_index = 0

for index in range(1, 6):
    if data_list[index] > max:
        max = data_list[index]
        max_index = index

print(max, max_index)
```

```python
data_list = [123, 45, 10, 14, 2, 67]

max = data_list[0]

for data in data_list:
    if data > max:
        max = data

print(max)
```

```python
# 리스트에 저장된 숫자 중 가장 작은 숫자 출력 (최소값)

data_list = [123, 45, 10, 14, 2, 67]

min = data_list[0]

for data in data_list:
    if data < min:
        min = data

print(min)
```

```python
# 리스트에 저장된 숫자 중 입력된 숫자보다 큰 수가 몇 개 있는지 구하기

data_list = [123, 45, 10, 14, 2, 67]
num = int(input())
numbers = 0

for i in data_list:
    if i > num:
        numbers += 1

print(numbers)
```

```python
# 리스트에 저장된 모든 값 더하여 출력(합)

data_list = [123, 45, 10, 14, 2, 67]
summary = 0

for i in data_list:
    summary += i

print(summary)
```

```python
# 리스트에 저장된 숫자를 역순으로 저장하여 출력

data_list = [123, 45, 10, 14, 2, 67]

for i in range(4, -1, -1):
    data_list.append(data_list[i])

for i in range(0, 5):
    data_list.pop(0)

print(data_list)
```
